O **readlines()** lê todas as linhas do arquivo de uma vez e as carrega na memória como uma lista. No código, ele foi usado para podermos aplicar o fatiamento [1:], ignorando a primeira linha de cabeçalho do arquivo CSV.

https://docs.python.org/pt-br/3.14/library/io.html

https://pt.stackoverflow.com/questions/437384/como-acessar-as-linhas-de-um-arquivo-lidas-por-readlines

```python
with open("vendas.csv", "r", encoding="utf-8") as arq:
    for linha in arq.readlines()[1:]:
        data, produto, qtd, valor = linha.strip().split(",")
        print(f"{produto}: {qtd} unidades - R${valor}")
```

```python
with open("vendas.csv", "r", encoding="utf-8") as arq:
    for i, linha in enumerate(arq):
        if i == 0:
            continue
        data, produto, qtd, valor = linha.strip().split(",")
        print(f"{produto}: {qtd} unidades - R${valor}")
```

```python
with open("vendas.csv", "r", encoding="utf-8") as arq:
    for num_linha, linha in enumerate(arq):
        if (1 <= num_linha <= 2) or (4 <= num_linha <= 5):
            data, produto, qtd, valor = linha.strip().split(",")
            print(f"{produto}: {qtd} unidades - R${valor}")


```

Se vocês quiserem, podem criar bases sintéticas para testar o código abaixo.  
É legal que seja uma base grande para verem a diferença do poder de processamento.

```python
import tracemalloc

# 1. Inicia o monitoramento de memória
tracemalloc.start()

# --- SEU CÓDIGO COMEÇA AQUI ---
pass
# --- SEU CÓDIGO TERMINA AQUI ---

# 2. Captura o consumo atual e o pico de memória
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

# 3. Exibe os resultados convertidos para Megabytes (MB)
print(f"Consumo atual: {current / 10**6:.2f} MB")
print(f"Pico de consumo da célula: {peak / 10**6:.2f} MB")
```

```python
import tracemalloc

tracemalloc.start()



with open("Base_Busca.csv", "r", encoding="utf-8") as arq:
    # next(arq) # a gente pode usar isso para pular a primeira linha
    for linha in arq:
        print(linha)


print('\n')



current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f"Consumo atual: {current / 10**6:.2f} MB")
print(f"Pico de consumo da célula: {peak / 10**6:.2f} MB")
```

```python
import tracemalloc

tracemalloc.start()





with open("Base_Busca.csv", "r", encoding="utf-8") as arq:
    for linha in arq.readlines()[1:]:
        print(linha)



print('\n')

current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()

print(f"Consumo atual: {current / 10**6:.2f} MB")
print(f"Pico de consumo da célula: {peak / 10**6:.2f} MB")
```

## 📊 Análise dos Dados

* Pico de consumo (xxx MB): Este foi o teto máximo de memória que a sua célula exigiu do computador.(um navegador web com duas abas abertas costuma gastar mais de 500 MB). 
* Consumo atual (xxxx MB): Esta é a memória que continuou presa no sistema após o código terminar de rodar.

## 💡 O Diagnóstico Principal
A diferença entre o pico e o atual revela o comportamento do código (em linhas gerais):

   1. O código não gerou lixo temporário: Ele não criou grandes matrizes ou variáveis intermediárias que precisaram ser apagadas depois.
   2. Os dados foram mantidos na memória: Quase tudo o que a célula processou (provavelmente um DataFrame do Pandas, uma lista ou um modelo) foi salvo em uma variável global e continua disponível para as próximas células utilizarem.
