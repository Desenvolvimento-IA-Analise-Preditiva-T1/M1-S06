## Funções lambdas - Exemplo 2

```python
lista_num = list(range(1,11))
print(lista_num)
```

```python
# O filter recebe: uma função que retorna Verdadeiro/Falso e a lista.
# A lambda verifica se o resto da divisão por 2 é zero (x % 2 == 0)

numeros_pares = list(filter(lambda x: x % 2 == 0, lista_num))

print(numeros_pares)
```
